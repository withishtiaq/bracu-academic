```python
#Ans to the question no: 1:

class Calculator:
    def __init__(self,num1,num2,operator):
        self.num1=num1
        self.num2=num2
        self.operator=operator
        print("Let's Calculate!")
    
    
    def namely_add(self):
        add= self.num1+self.num2
        print("Value 1:",self.num1)
        print("Operator:",self.operator)
        print("Value 2:",self.num2)
        print("Result:",add)
        
    def subtract(self):
        subtract=self.num1-self.num2
        print("Value 1:",self.num1)
        print("Operator:",self.operator)
        print("Value 2:",self.num2)
        print("Result:",subtract)
        
    def multiply(self):
        multiply = self.num1*self.num2
        print("Value 1:",self.num1)
        print("Operator:",self.operator)
        print("Value 2:",self.num2)
        print("Result:",multiply)
        
    def divid(self):
        divid = self.num1/self.num2
        print("Value 1:",self.num1)
        print("Operator:",self.operator)
        print("Value 2:",self.num2)
        print("Result:",divid)
    

num1=int(input("Enter your first number: "))
operator=input("Enter your operator")
num2=int(input("Enter your second number"))

c1=Calculator(num1,num2,operator)
if operator =="+":
    c1.namely_add()
elif operator=="-":
    c1.subtract()
elif operator == "*":
    c1.multiply()
elif operator =="/":
    c1.divid()
```

```python
#Ans to the question no. 2:

class Customer:
    def __init__(self,name):
        self.name=name
    def greet(self, name= None):
        if name!=None:
            print("Hello ",self.name+"!")
        else:
            print("Hello!")
    def purchase(self,*items):
        print(self.name,"you purchased",len(items),"item(s)")
        for i in items:
            print(i)
        
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")        
```

```python
#Ans to the question no. 9:

class Patient:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        self.symptoms=[]
    def add_Symptom(self,*s):
        for x in s:
            self.symptoms.append(x)
    def printPatientDetail(self):
        print("Name: ",self.name)
        print("Age: ",self.age)
        print("Symptoms:",end="")
        print(",".join(self.symptoms))
                 
        
p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")        
```

```python
#Ans to the question No 11:

class Shinobi:
    def __init__(self, name, rank):
        self.name = name
        self.rank = rank
        self.salary = 0
        self.mission = 0

    def changeRank(self, changeRank):
        self.rank = changeRank

    def calSalary(self, updateMission):
        self.mission = updateMission
        if self.rank == 'Genin':
            self.salary = self.mission * 50
        elif self.rank == 'Chunin':
            self.salary = self.mission * 100
        else:
            self.salary = self.mission * 500

    def printInfo(self):
        print('Name: ', self.name)
        print('Rank: ', self.rank)
        print('Number of mission: ', self.mission)
        print('Salary: ', self.salary)



naruto = Shinobi('Naruto', 'Genin')
naruto.calSalary(5)
naruto.printInfo()
print('=========================')
shikarmaru = Shinobi('Shikamaru', 'Genin')
shikarmaru.printInfo()
shikarmaru.changeRank('Chunin')
shikarmaru.calSalary(10)
shikarmaru.printInfo()
print('=========================')
neiji = Shinobi('Neiji', 'Jonin')
neiji.calSalary(5)
neiji.printInfo()
```

```python
#Ans to the question No.13:

class Batsman:
    def __init__(self, *args): 
        if isinstance(args[0], int):
            self.name = "New Batsman" 
            self.runs = args[0] 
            self.balls = args[1] 
        elif isinstance(args[0], str): 
            self.name = args[0] 
            self.runs = args[1] 
            self.balls = args[2] 
  
    def printCareerStatistics(self):
        print("Name:", self.name)
        print("Runs Scored:", self.runs, ", Balls Faced:", self.balls)
        
    def setName(self, name): 
        self.name = name
    
    def battingStrikeRate(self): 
        return (self.runs/self.balls) * 100
    
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("====================================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("-----------------")
print(b2.battingStrikeRate())
print("==================================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("--------------------")
print(b1.battingStrikeRate())
```

```python
#Ans to the question no. 14:

class EPL_Team:
    def __init__(self,name,*menu):
        self.name=name
        if len(menu)>0:
            self.song=menu[0]
        else:
            self.song='No Slogan'
        self.title=0

    def showClubInfo(self):
        return("Name: %s\nSong: %s\nTotal No of title: %d"%(self.name,self.song,self.title))
  
    def increaseTitle(self):
        self.title=self.title+1

    def changeSong(self,song):
        self.song=song



manu = EPL_Team('Manchester United','Glory Glory Man United')
chelsea=EPL_Team('Chelsea')


print('========================')
print(manu.showClubInfo())
print('########################')
manu.increaseTitle()
print(manu.showClubInfo())
print('========================')
print(chelsea.showClubInfo())
chelsea.changeSong('Keep the blue flag flying high')
print(chelsea.showClubInfo())
```

```python
#Ans to the question no. 15:

class Account:
    def __init__(self,name="Default Account",balance=0.0):
        self.name = name
        self.balance = balance

    def details (self):
        return str(self.name)+"\n"+str(self.balance)

    def withdraw(self, amount):
        if self.balance-amount>3070:
            self.balance = self.balance - amount
            print("Withdraw Successful! New balance is: "+str(self.balance))
        else:
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
            
a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930);
print("------------------------")
a2.withdraw(600);
print("------------------------")
a1.withdraw(6929)            
```

```python
#Ans to the question no 16:

class Author:
    
    def __init__(self, name="Default", *books):
        self.name = name
        self.books = []
        for book in books:
            self.books.append(book)

    def addBooks(self, *newbooks):
        for i in newbooks:
            self.books.append(i)
  
    def changeName(self, name):
        self.name = name
  
    def printDetails(self):
        print("Author Name:", self.name)
        print("-------")
        print("List of Books:")
        for book in self.books:
            print(book)

auth1 = Author("Humayun Ahmed")
auth1.addBooks("Deyal", "Megher Opor Bari")
auth1.printDetails()
print("======================")
auth2 = Author()
print(auth2.name)
auth2.changeName("Mario Puzo")
auth2.addBooks("The Godfather", "Omerta", "The Sicilian")
print("======================")
auth2.printDetails()
print("======================")
auth3 = Author("Paulo Coelho", "The Alchemist", "The Fifth Mountain")
auth3.printDetails()

```
