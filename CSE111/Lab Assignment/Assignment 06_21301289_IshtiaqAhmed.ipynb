```python
#Ans to the question no 8:


class Cylinder:
    radius = 5
    height = 18

    def __init__(self, radius, height):
        print("Default radius=",Cylinder.radius,"and height=",Cylinder.height)
        self.radius = radius
        self.height = height
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print("Updated: radius=",self.radius,"and height=",self.height)

    @staticmethod
    def area(radius, height):
        area = 2 *  3.141592653589793238* float(radius) * float(radius) + 2 *  3.141592653589793238 * float(radius) * float(height)
        print("Area:",area)

    @staticmethod
    def volume(radius, height):
        volume =  3.141592653589793238* float(radius) * float(radius) * float(height)
        print("Volume:",volume)

    @classmethod
    def swap(cls, radius, height):
        return Cylinder(height, radius)

    @classmethod
    def changeFormat(cls, string):
        radius, height = string.split('-')
        return Cylinder(float(radius), float(height))


c1 = Cylinder(0, 0)
Cylinder.area(c1.radius, c1.height)
Cylinder.volume(c1.radius, c1.height)
print("===============================")
c2 = Cylinder.swap(8, 3)
c2.area(c2.radius, c2.height)
c2.volume(c2.radius, c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius, c3.height)
c3.volume(c3.radius, c3.height)
print("===============================")
Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
print("===============================")
Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
```

```python
#Ans to the question no 10:

class SultansDine:
    lst = []
    name = ''
    Branch_sell = 0
    
    def __init__(self,name):
        self.name = name
        SultansDine.lst.append(self)

    def sellQuantity(self, quantity):
        if quantity<10:
            self.Branch_sell = quantity*300
        elif quantity<20:
            self.Branch_sell = quantity*350
        else:
            self.Branch_sell = quantity*400

    def branchInformation(self):
        print("Branch Name:",self.name)
        print("Branch Sell:",self.Branch_sell,"Taka")

    def details():
        lst = SultansDine.lst
        sell=0
        branchNo=len(lst)
        
        for item in lst:
            sell += item.Branch_sell

        print("Total number of branch(s):", branchNo)
        print("Total Sell:",sell,"Taka")
        
        if branchNo == 0:
            return

        for branch in lst:
            print("Branch Name:",branch.name,"Branch sell:",branch.Branch_sell,"Taka")
            sellP=round(branch.Branch_sell*100.0/sell,2)
            print("Branch consists of total sells:",sellP,"%")
            
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
